Importar las librerias necesarias para la manipulación de los datos. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Carga de Archivo cvs para el análisis de los datos

```python
data_co2 = pd.read_csv("../Emmision_CO2_Auto/DATA.csv", sep=",", encoding="utf-8")
data_co2.drop("Unnamed: 5", axis=1, inplace=True)
data_co2.info()
data_co2.head()
```

Ordenar por Marcas de Autos.

```python
data_co2 = data_co2.sort_values(by="Car")
data_co2.head()
```

Visualizar los datos

```python
sns.barplot(x="Model", y="CO2", data=data_co2, hue="Car")
plt.xticks(rotation=90, fontsize=9, ha="right")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```
